```python
datos_personas = []
nombre = "Señor/a"

while True:
    print("Menú:")
    print("1. Grabar")
    print("2. Buscar")
    print("3. Imprimir certificados")
    print("4. Eliminar")
    print("5. Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
      #RUN
        while True:
            run = input("Ingrese el RUN: ")
            run = run.replace("-", "").replace(".", "")
            if not run[:-1].isdigit() or run[-1] not in "0123456789Kk" or len(run) < 2:
                print("RUN inválido")
            elif len(run)<9:
                print("El RUN debe tener al menos 9 digitos, Incluido su digito verificador")
            else:
                break
      #FECHA NACIMIENTO
        while True:
            fecha = input("Ingrese la fecha de nacimiento (formato: DD/MM/AAAA): ")
            try:
               dia, mes, año = map(int, fecha.split('/'))
               if 1 <= dia <= 31 and 1 <= mes <= 12 and 1900 <= año <= 2023:
                break
               else:
                  print("Fecha inválida")
            except ValueError:
               print("Formato de fecha inválido")
      #NOMBRE
        while True:
            nombre = input("Ingrese el nombre: ")
            nombret = nombre.replace(" ", "")
            if not nombret.isalpha() or len(nombret) < 4 or len(nombret) > 40:
                print("El nombre debe contener solo letras y tener entre 4 y 20 caracteres")
            else:
                break
      #EDAD
        while True:
            edad = input("Ingrese su edad: ")
            if not edad.isdigit() or not (0 <= int(edad) <= 110):
                print("La edad debe ser numérica")
            else:
                break
      #Estado Civil
        while True:
            estado_civil = input("Ingrese el estado civil (soltero, casado, viudo, divorciado): ")
            estado_civil = estado_civil.lower()

            if estado_civil == "soltero" or estado_civil == "soltera":
              estado_civil = "Soltero/a"
              break
            elif estado_civil == "casado" or estado_civil == "casada":
              estado_civil = "Casado/a"
              break
            elif estado_civil == "viudo" or estado_civil == "viuda":
              estado_civil = "Viudo/a"
              break
            elif estado_civil == "divorciado" or estado_civil == "divorciada":
              estado_civil = "Divorciado/a"
              break
            else:
                print("Estado civil inválido. Por favor, ingrese uno válido (soltero, casado, viudo, divorciado).")
      #Pais Nacimiento
        while True:
            pais_nacimiento = input("Ingrese el país de nacimiento: ")
            if not pais_nacimiento.isalpha() or len(pais_nacimiento) < 4 or len(pais_nacimiento) > 20:
                print("El país de nacimiento debe contener solo letras y tener entre 4 y 20 caracteres")
            else:
                break
      #Ciudad Nacimiento
        while True:
            ciudad_nacimiento = input("Ingrese la ciudad de nacimiento: ")
            ciudad_nacimientot = ciudad_nacimiento.replace(" ", "")
            if not ciudad_nacimientot.isalpha() or len(ciudad_nacimientot) < 4 or len(ciudad_nacimientot) > 20:
                print("La ciudad de nacimiento debe contener solo letras y tener entre 4 y 20 caracteres")
            else:
                break
      #Datos guardados de la persona ingresada
        persona = {
            "run": run,
            "nombre": nombre,
            "fecha_nacimiento" : {
                "dia":dia,"mes":mes,"año":año
                },
            "estado_civil" : estado_civil,
            "edad": int(edad),
            "pais_nacimiento": pais_nacimiento,
            "ciudad_nacimiento": ciudad_nacimiento
        }

        datos_personas.append(persona)
        print("Datos de persona guardados exitosamente")

    elif opcion == "2":
        run = input("Ingrese el RUN de la persona a buscar: ")
        run = run.replace(".", "").replace("-", "")
        if not run[:-1].isdigit() or run[-1] not in "0123456789Kk" or len(run) < 2:
            print("RUN inválido")
            continue

        persona_encontrada = None
        for persona in datos_personas:
            if persona["run"] == rut:
                persona_encontrada = persona
                break

        if persona_encontrada:
            print("Información de la persona:")
            print("RUN:", persona_encontrada["run"])
            print("Nombre:", persona_encontrada["nombre"])
            print("Fecha de Nacimiento: {}/{}/{}".format(
                  persona_encontrada["fecha_nacimiento"]["dia"],
                  persona_encontrada["fecha_nacimiento"]["mes"],
                  persona_encontrada["fecha_nacimiento"]["año"]))
            print("Edad:", persona_encontrada["edad"])
            print("Estado Civil", persona_encontrada["estado_civil"])
            print("País de Nacimiento:", persona_encontrada["pais_nacimiento"])
            print("Ciudad de Nacimiento:", persona_encontrada["ciudad_nacimiento"])
            if persona_encontrada["pais_nacimiento"].lower() == "chile":
                print("La persona pertenece a Chile")
            elif pais_nacimiento:
                print("La persona pertenece a ",pais_nacimiento)
            else:
                print("No se encontro el pais de nacimiento")
        else:
            print("No se encontró ninguna persona con ese RUT")

    elif opcion == "3":
        run = input("Ingrese el RUT de la persona para imprimir sus certificados: ")
        run = run.replace("-", "")
        if not run[:-1].isdigit() or run[-1] not in "0123456789Kk" or len(run) < 2:
           print("RUN inválido")
           continue

        persona_encontrada = None
        persona_encontradaf = False

        for persona in datos_personas:
          if persona["run"] == run:
              persona_encontrada = persona
              break

        if persona_encontrada:
            print("Certificados de la persona con RUN:", persona_encontrada["run"])
            print("Nombre:", persona_encontrada["nombre"])
            persona_encontradaf = True

        if persona_encontradaf:
            while True:
              print("\nSubmenú de Certificados:")
              print("1. Certificado de Nacimiento")
              print("2. Certificado de Estado Conyugal")
              print("3. Certificado de Pertenencia")
              print("4. Volver al Menú Principal")

              subopcion = input("Seleccione una opción: ")

              if subopcion == "1":
                  print("")
                  print("========================================================")
                  print("\t\tCertificado de Nacimiento")
                  print("========================================================")
                  print("")
                  print("CIRCUNSCRIPCIÓN:\t",ciudad_nacimiento.upper())
                  print("AÑO DE INSCRIPCIÓN:\t",año)
                  print("NOMBRE:\t\t\t",nombre.upper())
                  print("RUN:\t\t\t",rut)
                  print("")
                  print("========================================================")
                  print("\t\tCertificado Digital Gratuito")
                  print("\t\t   www.registrocivil.cl")
                  print("========================================================")
                  print("")

              elif subopcion == "2":
                  print("")
                  print("========================================================")
                  print("\t\tCertificado de Estado Civil")
                  print("========================================================")
                  print("")
                  print("NOMBRE:\t\t\t",nombre.upper())
                  print("RUN:\t\t\t",rut)
                  print("ESTADO CIVIL\t\t",estado_civil.upper())
                  print("")
                  print("========================================================")
                  print("\t\tCertificado Digital Gratuito")
                  print("\t\t   www.registrocivil.cl")
                  print("========================================================")
                  print("")

              elif subopcion == "3":
                  print("")
                  print("========================================================")
                  print("\t\tCertificado de Pertenencia")
                  print("========================================================")
                  print("")
                  print("NOMBRE:\t\t\t",nombre.upper())
                  print("RUN:\t\t\t",run)
                  print("CIUDAD DE NACIMIENTO:\t",ciudad_nacimiento.upper())
                  print("PAIS DE NACIMIENTO:\t", pais_nacimiento.upper())
                  print("")
                  print("SE ACREDITA QUE LA PERSONA")
                  print("")
                  print("NOMBRE:\t\t\t",nombre.upper(),)
                  print("PERTENECE A:\t\t",pais_nacimiento.upper())
                  print("")
                  print("========================================================")
                  print("\t\tCertificado Digital Gratuito")
                  print("\t\t   www.registrocivil.cl")
                  print("========================================================")
                  print("")

              elif subopcion == "4":
                print("Volviendo al Menú Principal.")
                break

              else:
                print("Opción inválida")

        else:
            print("No se encontró ninguna persona con ese RUN")



    elif opcion == "4":
        run = input("Ingrese el RUN de la persona a eliminar: ")
        run = run.replace("-", "")
        if not run[:-1].isdigit() or run[-1] not in "0123456789Kk" or len(run) < 2:
            print("RUN inválido")
            continue

        persona_eliminada = None
        for i, persona in enumerate(datos_personas):
            if persona["run"] == run:
                persona_eliminada = datos_personas.pop(i)
                break

        if persona_eliminada:
            print("Datos de persona eliminados exitosamente")
        else:
            print("No se encontró ninguna persona con ese RUN")

    elif opcion == "5":
        print(f"¡Hasta luego! {nombre}\n¡Gracias por utilizar Nuestros Servicios!\n¡Versión de Sistema 1.11!")
        break

    else:
        print("Opción inválida")
```
